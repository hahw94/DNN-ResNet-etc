```python
used_column = ['Month', 'Aerosol classification', 'Temperature', 'Local Pressure', 'SNU_PBLH', 'Dewpoint temperature', 'Wind Direction', 'Humidity', 'Total cloud cover', 'AOD', 'Single Scattering Albedo ', 'NDVI', 'Wind Speed','PM10(Total)']
```

```python
result = df[used_column]
```

```python
for column in used_column:
    result[column] = result[result != -999][column].interpolate().dropna()
```

```python
result.dropna()
```

```python
result.mean()
```

```python
result.std()
```

```python
mean = result['PM10(Total)'].mean()
std = result['PM10(Total)'].std()
```

```python
result_2 = (result - result.mean()) / result.std()
result_2.dropna().count()
```

```python
result_2_columns = result_2.columns.difference(["Aerosol classification","Total cloud cover"])
```

```python
final = result_2[result_2_columns].dropna()
```

```python
train_data = final[final.columns.difference(["PM10(Total)"])]
```

```python
train_data.head()
```

```python
train_data.to_csv("train_test_data.csv")
```

```python
target_data = (final['PM10(Total)'] * std) + mean
```

```python
target_data.head()
```

```python
target_data.to_csv("target_data.csv")
```

```python
train_data_numpy = train_data.to_numpy()
target_data_numpy = target_data.to_numpy()
```
